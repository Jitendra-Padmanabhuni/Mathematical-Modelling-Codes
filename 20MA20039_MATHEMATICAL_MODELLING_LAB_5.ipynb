```python
#1
x_0=1
y_0=1
x=[]
x.append(x_0)
y=[]
y.append(y_0)
d_0=(x_0**2+y_0**2)**0.5
d=[]
d.append(d_0)
t=[0]
h=float(input('give step length: '))
T=float(input('give simulation time: '))
for i in range(int(T/h)):
    x.append(x[i]+h*(2*x[i]+2*y[i]+3*t[i]))
    y.append(y[i]+h*(x[i]+y[i]+(t[i]/2)))
    t.append(t[i]+h)
    d.append((x[i]**2+y[i]**2)**0.5)
import numpy as np
x=np.array(x)
d=np.array(d)
y=np.array(y)
t=np.array(t)
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.xlabel('x component')
plt.ylabel('y component')
plt.title('motion of particle')
plt.show()
plt.plot(t,x)
plt.xlabel('t')
plt.ylabel('x')
plt.title('x vs t')
plt.show()
plt.plot(t,y)
plt.xlabel('t')
plt.ylabel('y')
plt.title('y vs t')
plt.show()
plt.plot(t,d)
plt.xlabel('time')
plt.ylabel('displacement')
plt.title('displacement vs time')
plt.show()
```
