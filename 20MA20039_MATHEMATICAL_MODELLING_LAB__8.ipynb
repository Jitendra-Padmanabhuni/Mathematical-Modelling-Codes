```python
#1
import math
def u_0(x):
    return (0.25*math.sin(x))
x_high=1
t_high=10
step_x=0.1
step_t=0.5
c=1
x=[]
x.append(0)
t=[]
t.append(0)
u=[[]]
for i in range(int(x_high/step_x)):
    x.append(x[i]+step_x)
for i in range(int(t_high/step_t)):
    t.append(t[i]+step_t)
for i in x:
    u[0].append(u_0(i))
for n in range(len(t)-1):
    u_new=[]
    for i in range(len(x)-1):
        u_new.append(u[n][i+1]*(step_t/(step_x*c))+u[n][i]*(1-(step_t/(step_x*c))))
    u_new.append(0)
    u.append(u_new)
u_points=[]
for n in range(len(t)):
    for i in range(len(x)):
        u_temp=[]
        u_temp.append(u[n][i])
        u_temp.append(x[i])
        u_temp.append(t[n])
        u_points.append(u_temp)
x_data=[]
t_data=[]
u_data=[]
for i in range(len(u_points)):
    x_data.append(u_points[i][1])
    t_data.append(u_points[i][2])
    u_data.append(u_points[i][0])
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
ax=plt.axes(projection="3d")
x_data=np.array(x_data)
t_data=np.array(t_data)
u_data=np.array(u_data)
ax.scatter(x_data,t_data,u_data)
plt.show()
```
