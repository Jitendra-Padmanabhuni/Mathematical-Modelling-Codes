```python
#spruce-budworm model
r=float(input('give r: '))
k=float(input('give k: '))
B=float(input('give B: '))
A=float(input('give A: '))
y_0=float(input("give initial population: "))
t=float(input("give time period: "))
h=float(input("step length: "))
x = [0]
y=[]
y.append(y_0)
for i in range(int(t/h)):
    y.append((y[i]+r*y[i]*(1-(y[i]/k))*h)-(B*(y[i]**2)/((A**2)+(y[i]**2))))
    x.append(x[i]+h)
import numpy as np
x = np.array(x)
y = np.array(y)
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('spruce budworm growth plot')
plt.show()
```

```python
#prey-predator model
a1=float(input('give alpha1: '))
k=float(input('give k: '))
B=float(input('give beta: '))
a2=float(input('give alpha2: '))
g=float(input('give gamma: '))
y_0=float(input("give initial prey population: "))
z_0=float(input('give initial predator poluation: '))
t=float(input("give time period: "))
h=float(input("step length: "))
x = [0]
y=[]
y.append(y_0)
z=[]
z.append(z_0)
for i in range(int(t/h)):
    y.append(y[i]+h*(a1*y[i]*(1-(y[i]/k)-(B*z[i]))))
    z.append(z[i]+h*((-1)*a2*z[i]*(1-g*y[i])))
    x.append(x[i]+h)
import numpy as np
x = np.array(x)
y = np.array(y)
z = np.array(z)
import matplotlib.pyplot as plt
plt.plot(x,y,color='r',label='prey')
plt.plot(x,z,color='g',label='predator')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('interacting species growth plot')
plt.show()
```

```python
#modified predator-prey model
a=float(input('give alpha: '))
B=float(input('give beta: '))
y_0=float(input("give initial prey population: "))
z_0=float(input('give initial predator population: '))
t=float(input("give time period: "))
h=float(input("step length: "))
x = [0]
y=[]
y.append(y_0)
z=[]
z.append(z_0)
for i in range(int(t/h)):
    y.append(y[i]+h*((a*y[i])-(B*z[i])-y[i]*(y[i]**2 + z[i]**2)))
    z.append(z[i]+h*((B*y[i])+(a*z[i])-z[i]*(y[i]**2 +z[i]**2)))
    x.append(x[i]+h)
import numpy as np
x = np.array(x)
y = np.array(y)
z = np.array(z)
import matplotlib.pyplot as plt
plt.plot(x,y,color='r',label='prey')
plt.plot(x,z,color='g',label='predator')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('modified interacting species growth plot')
plt.show()
```
