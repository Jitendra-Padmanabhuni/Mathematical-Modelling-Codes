```python
#cancer-tumor-immune system modelling
lc=0.01
uc=0.00001
c0=1000000
u=0.3
k11=3000
g=200
kt=3300
k1=50000
k2=100000
ut=0.2
h=0.1
t=100
c_0=100
m1_0=50000
m2_0=0
T_0=0
x = [0]
c=[]
c.append(c_0)
m1=[]
m1.append(m1_0)
m2=[]
m2.append(m2_0)
T=[]
T.append(T_0)
for i in range(int(t/h)):
    c.append(c[i]+h*(lc*c[i]*(1-(c[i]/c_0))-uc*T[i]*c[i]))
    m1.append(m1[i]+h*(k11-g*m1[i]*(c[i]/(k1+c[i]))-u*m1[i]))
    m2.append(m2[i]+h*(g*m1[i]*(c[i]/(k1+c[i]))-u*m2[i]))
    T.append(T[i]+h*(kt*(m1[i]/(k2+m2[i]))-ut*T[i]))
    x.append(x[i]+h)
import numpy as np
x = np.array(x)
c = np.array(c)
m1 = np.array(m1)
m2 = np.array(m2)
T = np.array(T)
import matplotlib.pyplot as plt
plt.plot(x,c,label='cancer cell')
plt.plot(x,m1,label='m1 cell')
plt.plot(x,m2,label='m2 cell')
plt.plot(x,T,label='T cell')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('cancer-tumor-immune system growth plot')
plt.show()
```
