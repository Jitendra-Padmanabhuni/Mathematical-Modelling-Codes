```python
#continuous growth model
l=float(input('give lambda: '))
y_0=float(input("give initial population: "))
t=float(input("give time period: "))
h=float(input("step length: "))
x = [0]
y=[]
y.append(y_0)
for i in range(int(t/h)):
    y.append(y[i]+l*x[i]*h)
    x.append(x[i]+h)
import numpy as np
x = np.array(x)
y = np.array(y)
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('continuous population growth plot')
plt.show()
```

```python
#logistic growth model
r=float(input('give r: '))
k=float(input('give k: '))
y_0=float(input("give initial population: "))
t=float(input("give time period: "))
h=float(input("step length: "))
x = [0]
y=[]
y.append(y_0)
for i in range(int(t/h)):
    y.append(y[i]+r*y[i]*(1-(y[i]/k))*h)
    x.append(x[i]+h)
import numpy as np
x = np.array(x)
y = np.array(y)
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('logistic population growth plot')
plt.show()
```
