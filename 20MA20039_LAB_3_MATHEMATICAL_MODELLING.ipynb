```python
#kermack-mckendrick model
r=float(input('give r: '))
a=float(input('give a: '))
s_0=float(input('give initial susceptible poulation: '))
i_0=float(input('give initial infected population: '))
r_0=float(input('give initial population of people who are dead or recovered: '))
t=float(input("give time period: "))
h=float(input("step length: "))
x = [0]
S=[]
R=[]
I=[]
R.append(r_0)
I.append(i_0)
S.append(s_0)
for i in range(int(t/h)):
    S.append(S[i]-h*(r*I[i]*S[i]))
    I.append(I[i]+h*((r*I[i]*S[i])-(a*I[i])))
    R.append(R[i]+h*(a*I[i]))
    x.append(x[i]+h)
import numpy as np
x = np.array(x)
S = np.array(S)
R = np.array(R)
I = np.array(I)
import matplotlib.pyplot as plt
plt.plot(x,S,color='red',label='susceptible population')
plt.plot(x,I,color='blue',label='infected infected')
plt.plot(x,R,color='green',label='remaining population')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('kermack-mckendrick epidemic population vs time')
plt.show()
plt.plot(I,S)
plt.xlabel('infected population')
plt.ylabel('susceptible poulation')
plt.show()
plt.plot(R,S)
plt.xlabel('recovered population')
plt.ylabel('susceptible poulation')
plt.show()
```

```python
#kermack-mckendrick model with demography
r=float(input('give r: '))
a=float(input('give a: '))
b=float(input('birth rate: '))
d=float(input('death rate: '))
s_0=float(input('give initial susceptible poulation: '))
i_0=float(input('give initial infected population: '))
r_0=float(input('give initial population of people who have recovered: '))
N=s_0+i_0+r_0
t=float(input("give time period: "))
h=float(input("step length: "))
x = [0]
S=[]
R=[]
I=[]
R.append(r_0)
I.append(i_0)
S.append(s_0)
for i in range(int(t/h)):
    S.append(S[i]-h*(d*S[i]-b*N+r*I[i]*S[i]/N))
    I.append(I[i]+h*((r*I[i]*S[i]/N)-(a*I[i])-d*I[i]))
    R.append(R[i]+h*(a*I[i]-d*R[i]))
    x.append(x[i]+h)
import numpy as np
x = np.array(x)
S = np.array(S)
R = np.array(R)
I = np.array(I)
import matplotlib.pyplot as plt
plt.plot(x,S,color='red',label='susceptible population')
plt.plot(x,I,color='blue',label='infected infected')
plt.plot(x,R,color='green',label='remaining population')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('kermack-mckendrick epidemic population vs time')
plt.show()
```
